```python
import matplotlib.pyplot as plt
import numpy as np
import math

x= np.arange(-10,11)
y= x**2

plt.plot(x,y)
plt.show()
```

```python
x= np.arange(-10,11)
y= -x**2 + 100
plt.plot(x,y)
plt.show()
```

```python
x= np.arange(0,10)
y= []

for i in x:
    y.append(math.exp(i))

plt.plot(x,y)
plt.show()

```

```python
x = np.arange(-10,11)

y = []
for i in x:
    y.append(np.random.randint(1,10))

plt.step(x,y)
plt.show()

```

```python
x = np.arange(-5,6)

f_x = []

for i in x:
    f_x.append(1/(1+math.e**-i))


plt.plot(x,f_x)
plt.show()
    
```

```python
x= np.arange(0,100000)
y= []

for i in x:
    y.append(-math.sqrt(i))

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-100,101)

y = -abs(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-100,101)

y = abs(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-0,101)

y = np.log(x)

plt.plot(x,y)
plt.show()
```

```python
np.seterr(divide='ignore', invalid='ignore')
x= np.arange(-100,101)

y= 1/(x)

plt.plot(x,y)
plt.show()
```

```python
x= np.arange(-100,101)

y= 1/(x**2)

plt.plot(x,y)
plt.show()
```
