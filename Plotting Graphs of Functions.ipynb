```python
x = np.arange(-100,101)

f_x = []

for i in x:
    if  -100<= i <= 2:
        f_x.append(i**2)
    if 2 < i <= 60:
        f_x.append(i**3)
    if i > 60:
        f_x.append(2*i + 1)
        
plt.plot(x,f_x)
plt.show()
    
```

```python
x = np.arange(-100,101)

f_x = []

for i in x:
    if  -100<= i <= 2:
        f_x.append(i**2)
    if 2 < i <= 60:
        f_x.append(i**3)
    if i > 60:
        f_x.append(-i**3)
        
plt.plot(x,f_x)
plt.show()
    
```
