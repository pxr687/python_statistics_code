```python
x = np.linspace(0,10,1000)

y = [] 
y1 = []
y2 = []
y3 = []

for i in x:
    y.append(2*math.exp(i))
    y1.append(8*math.exp(i))
    y2.append(32*math.exp(i))
    y3.append(64*math.exp(i))

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```
