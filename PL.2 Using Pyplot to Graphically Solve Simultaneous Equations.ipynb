```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(-4,4,1000)

y = (4-(3*x))/2
y1 = x - 3

plt.plot(x,y)
plt.plot(x,y1)
plt.grid()
plt.plot(2,-1, 'o', color = 'red')
plt.show()
```
