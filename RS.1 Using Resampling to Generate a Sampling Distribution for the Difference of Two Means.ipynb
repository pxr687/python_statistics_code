```python
import numpy as np
import pandas as pd
import statistics
import random
import matplotlib.pyplot as plt

# set the size of the simulated population(s)

pop_size = 100000

# set the size of the samples to be drawn

sample_size = 80

# set the number of times to iterate the entire resampling procedure

resample_iterations = 10

# set the number of times to resample on each iteration

resample_no = 1000

population = []
population2 = []

# making two populations of scores on a quantitative-discrete variable and generating summary statistics.
# the populations have different means.

print('Generating two populations of', pop_size, 'observations:')
print()

for i in range(pop_size):
    population.append(np.random.normal(loc = 100, scale = 25))
    population2.append(np.random.normal(loc = 150, scale = 25))
    
population = pd.DataFrame({'population_x_values': population})
population2 = pd.DataFrame({'population2_x_values': population2})

plt.figure(1, figsize = (14,4))
plt.hist([population['population_x_values'], population2['population2_x_values']], 
         label = ['population 1', 'population 2'], bins =50)
plt.legend()
plt.title('Frequency Distribution for Two Populations')
plt.xlabel('Score on Variable')
plt.ylabel('Frequency')
plt.show()

print('Population 1 Descriptive Statistics:', population.describe())
print()
print('Population 2 Descriptive Statistics:', population2.describe())

pop_mean_difference = np.mean(population2['population2_x_values']) - np.mean(population['population_x_values'])

print()
print('The difference between the means of the two populations is:', pop_mean_difference)

# generating the actual sampling distribution of the difference in means from the population(s)

print()
print('############################################################################################################################')
print()

print('On each iteration either:')
print()
print('a) Two samples will be drawn from population 1 (n =', sample_size,'for each sample).')
print()
print('OR')
print()
print('b) One sample will be drawn from population 1 and one sample will be drawn from population 2 (n =', sample_size,'for each sample).')
print()
print('The null hypothesis (H0) is that the samples were drawn from the same population.')
print()


for i in range(resample_iterations):
    print('############################################################################################################################')
    print()
    print('Iteration number:', i)
    print()
    # drawing samples either from the sample population or from different populations

    coinflip = np.random.choice(['heads', 'tails'])

    if coinflip == 'heads':
        sample1 = population['population_x_values'].sample(n=30)
        sample2 = population['population_x_values'].sample(n=30)
        print('On this iteration both of the samples were drawn from population 1. H0 is true.')
        print()
    else:
        sample1 = population['population_x_values'].sample(n=30)
        sample2 = population2['population2_x_values'].sample(n=30)
        print('On this iteration one sample was drawn from population 1, the other was drawn from population 2. H0 is false.')
        print()

    samp_mean_difference = np.mean(sample2)-np.mean(sample1)

    print('The difference in the sample means was:', samp_mean_difference)
    print()
    
    # pooling the samples and generating a pseudopopulation through randoming sampling

    pooled = sample1.append(sample2)

    simulated_sampling_dist = []

    for i in range(resample_no):
        simulated_sample1 = pooled.sample(n=30)
        simulated_sample2 = pooled.sample(n=30)
        simulated_sampling_dist.append(np.mean(simulated_sample1)-np.mean(simulated_sample2))

    simulated_sampling_dist = pd.DataFrame({'difference_in_means': simulated_sampling_dist})

    print('Here is the sampling distribution of the difference in means generated by pooling the samples and resampling them', resample_no, 'times in accordance with H0 (which states that the samples are drawn from the same population):')

    plt.figure(2, figsize = (14,6))
    plt.hist(simulated_sampling_dist['difference_in_means'], bins = 20, color = 'green', label = 'simulated sampling distribution (resampling according to H0)')
    plt.plot((samp_mean_difference),0, 'o', color = 'orange', 
             label = 'mean difference between original samples', markersize = 18)
    plt.title('Simulated Sampling Distribution of the Difference in Sample Means, created by Resampling')
    plt.xlabel('Difference in Means Between Two Samples')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
    print()
    
    # generating a null sampling distribution from the actual population to see how similar the resampled distribution is

    null_sample_dist = []
    for i in range(1000):

        sample1 = population['population_x_values'].sample(n=30)
        sample2 = population['population_x_values'].sample(n=30)
        null_sample_dist.append(np.mean(sample1)-np.mean(sample2))
    
        
    null_sample_dist = pd.DataFrame({'difference_in_means': null_sample_dist})
    
    plt.figure(3, figsize = (14,6))
    plt.hist([null_sample_dist['difference_in_means'],simulated_sampling_dist['difference_in_means']],
             label = ['actual H0 sampling distribution generated from population 1', 'sampling distribution generated through resampling two samples in accordance with H0'])
    plt.xlabel('Difference in Means Between Two Samples')
    plt.ylabel('Frequency')
    plt.legend(loc = 'upper left')
    plt.show()
    
    # calculating Cohen's D for the actual sampling distribution and the resampled one
    
    null_dist_mean_diff = np.mean(null_sample_dist['difference_in_means'])
    simulated_null_dist_mean_diff = np.mean(simulated_sampling_dist['difference_in_means'])

    mean_diff = abs(simulated_null_dist_mean_diff - null_dist_mean_diff) 

    pooled_sd = (np.sqrt((statistics.stdev(null_sample_dist['difference_in_means']) ** 2 + statistics.stdev(simulated_sampling_dist['difference_in_means']) ** 2) / 2))   

    cohens_d = mean_diff/pooled_sd
    
    print("Cohen's d was calculated to measure how well the sampling distribution generated by resampling approximated the actual sampling distribution. (A large value indicates a poor approximation, a small value indicates a good approximation):")
    print()
    print("Cohen's d between the actual sampling distribution and the resampled sampling distribution =", cohens_d)
    print()
    print("Cohen's d guide: small (d = 0.2), medium (d = 0.5), and large (d = 0.8) (see https://www.frontiersin.org/articles/10.3389/fpsyg.2013.00863/full)")
    print()
    print('The difference in the means between the resampled sampling distribution and the actual sampling distribution was:', mean_diff)
    print()
```
