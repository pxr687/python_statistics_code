```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import scipy
from pylab import rcParams

# setting a default figure size for timeseries plots

rcParams['figure.figsize'] = 12,4
```

```python
x = np.linspace(0,60, 10000)

y = np.sin(x)

df = pd.DataFrame({'time': x, 'amplitude': y})

plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
x = np.linspace(0,60, 10000)

y = scipy.signal.sawtooth(x)

df = pd.DataFrame({'time': x, 'amplitude': y})

plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
x = np.linspace(0,60, 10000)

y = scipy.signal.sawtooth(x, width = 0.5)

df = pd.DataFrame({'time': x, 'amplitude': y})

plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
x = np.linspace(0,60, 10000)

y = scipy.signal.square(x)

df = pd.DataFrame({'time': x, 'amplitude': y})

plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
```
